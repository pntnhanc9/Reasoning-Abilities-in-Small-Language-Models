## 2. Setup Paths and Extract Dataset

```python
import os
import zipfile
from pathlib import Path

# Kaggle paths
INPUT_DIR = Path("/kaggle/input")
WORKING_DIR = Path("/kaggle/working")
DATASET_DIR = INPUT_DIR / "self-steering-dataset"
OUTPUT_DIR = WORKING_DIR / "results"
OUTPUT_DIR.mkdir(exist_ok=True)

# If dataset is zipped, extract it
if (DATASET_DIR / "self-steering.zip").exists():
    with zipfile.ZipFile(DATASET_DIR / "self-steering.zip", 'r') as zip_ref:
        zip_ref.extractall(WORKING_DIR)
    PROJECT_DIR = WORKING_DIR / "self-steering"
else:
    PROJECT_DIR = DATASET_DIR

print(f"Project dir: {PROJECT_DIR}")
print(f"Output dir: {OUTPUT_DIR}")
```

## 3. Run the Benchmark

```python
# Change to project directory
os.chdir(PROJECT_DIR / "true_working")

# Run the benchmark script
!python run_benchmark.py --num-tasks 2 --results-dir /kaggle/working/results

print("Benchmark completed!")
```

## 4. Check Results

```python
# List output files
for file in OUTPUT_DIR.rglob("*"):
    if file.is_file():
        print(file)

print("\nResults saved to /kaggle/working/results/")
```
